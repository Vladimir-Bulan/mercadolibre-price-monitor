## ğŸ’¾ 4. Inicializar Base de Datos

Creamos la base de datos SQLite que usaremos para almacenar el histÃ³rico de precios.

```python
# Crear instancia de la base de datos
db = PriceDatabase("../data/prices.db")

print("Base de datos inicializada")
print("UbicaciÃ³n: ../data/prices.db\n")

# Verificar estadÃ­sticas
stats = db.get_stats()

print("ğŸ“Š EstadÃ­sticas de la base de datos:")
print("="*50)
print(f"Total de productos: {stats.get('total_products', 0)}")
print(f"Total de precios registrados: {stats.get('total_prices', 0)}")

if stats.get('first_record'):
    print(f"Primer registro: {stats['first_record']}")
    print(f"Ãšltimo registro: {stats['last_record']}")
else:
    print("\n(Base de datos vacÃ­a - empezando desde cero)")
```

## âœ… 5. Guardar Producto de Prueba

```python
if products:
    # Tomar el primer producto como prueba
    test_product = products[0]
    
    print("Guardando producto de prueba...\n")
    print_product_summary(test_product)
    
    # Guardar en la base de datos
    success = db.save_price(test_product)
    
    if success:
        print("âœ“ Producto guardado exitosamente en la base de datos")
        
        # Verificar que se guardÃ³
        history = db.get_price_history(test_product['id'])
        print(f"\nâœ“ Registros en el histÃ³rico: {len(history)}")
    else:
        print("âœ— Error al guardar el producto")
else:
    print("âš ï¸ No hay productos para guardar")
```
